```python
import random
import secrets


def GenerarLlave(n):
     return secrets.token_bytes(n) 

def E(k,m):
  c=[]
  for i in range(len(k)):
      ci=k[i]^m[i]  # ^ XOR.  11110101 ^ 00001111=11111010
      c.append(ci)
  return bytearray(c)

def D(k,c):
  m=[]
  for i in range(len(k)):
      mi=k[i]^c[i]
      m.append(mi)
  return bytearray(m)


Mensaje='Ricardo Villanueva Polanco'
M= bytes(Mensaje, 'ascii')

print(M)
K=GenerarLlave(len(M))
C=E(K,M)


#MP=b'2'*len(M)


#C=E(MP,C)

print(C.hex())

M=D(K,C)
print(str(M))
```
